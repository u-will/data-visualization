```python
import matplotlib.pyplot as plt
```

```python

import seaborn as sns
```

```python
sns.set()
```

```python
def plot_lines():
    plt.figure()
    xa = [10,20,5,40,8]
    xb = [30,43,9,7,20]
    plt.plot(xa, label="Spam")
    plt.plot(xb, label="Eggs")
    plt.legend()
    plt.show()

plot_lines()
```

```python
sns.load_dataset("car_crashes")
```

```python
sns.load_dataset("car_crashes").head(5)
```

```python
sns.relplot(x="alcohol", y="speeding", hue="speeding", data=sns.load_dataset("car_crashes"))
```

```python
sns.relplot(x="alcohol", y="total", data=sns.load_dataset("car_crashes").groupby("alcohol", as_index=False).sum())
sns.lmplot(x="alcohol", y="total", data=sns.load_dataset("car_crashes").groupby("alcohol", as_index=False).sum())
```
